## Pandas

A partir do dataset `movies.csv` utilize o Pandas para responder as seguintes perguntas:

```python
import pandas as pd

arquivo = "movies.csv"

table = pd.read_csv(arquivo)
table.head()
```

### Qual o gênero de filme mais lucrativo em todo o mundo?

```python
# ADICIONE O SEU CÓDIGO AQUI
lista = [table[table['Major Genre'].unique()[i] == table['Major Genre']]['Worldwide Gross'].sum() for i in range(len(table['Major Genre'].unique()))]
lista
```

```python
df_sum = pd.Series(lista, table['Major Genre'].unique())
df_sum.sort_values(ascending=False)
```

### Qual o diretor dirigiu mais filmes?

```python
# ADICIONE O SEU CÓDIGO AQUI
table["Director"].value_counts()
```

### Quais filmes tiverem os maiores arrecadações (Gross) em relação ao orçamento de produção (Budget)?

```python
# ADICIONE O SEU CÓDIGO AQUI
Gross = table.T.loc["Title":"Production Budget"].T
Gross = Gross.drop(columns="US DVD Sales")
Gross
```

```python
Gross["JungleDiff"] = Gross["Worldwide Gross"] - Gross["Production Budget"]
```

```python
Gross.sort_values(by="JungleDiff", ascending=False)
```
